```python
from os import listdir
from os.path import isfile, isdir, join
from PIL import Image 
  
def walk_through_file(path):
    files = listdir(path)
    for f in files:
        fullpath = join(path, f)
        if isfile(fullpath):
            image = Image.open(fullpath)
            width, height = image.size 
            print("------ f:%10s, width: %5d, height:%5d" %(f, width, height))
        elif isdir(fullpath):
            print("Directoryï¼š", f)
            walk_through_file(fullpath)
            
train_data_path = "./food/training"
test_data_path = "./food/evaluation"

#walk_through_file(train_data_path)
walk_through_file(test_data_path)
```
